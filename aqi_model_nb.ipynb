***Plotting histograms to check the distribution of pollutant concentration***

```python
# Importing the pyplot module from matplotlib for creating visualizations.
import matplotlib.pyplot as plt

# Importing the seaborn library for statistical data visualization, which works well with pandas.
import seaborn as sns

# Defining a function 'histograms' that takes a DataFrame 'df' as an argument.
def histograms(df):
    # Defining a color map to associate specific colors with different pollutants.
    color_map = {
        'PM2.5': 'blue', 'PM10': 'green', 'NO': 'red', 'NO2': 'purple',
        'NOx': 'orange', 'NH3': 'brown', 'CO': 'black', 'SO2': 'gray',
        'O3': 'teal', 'Benzene': 'pink', 'Toluene': 'gold', 'Xylene': 'cyan',
        'AQI': 'violet'
    }
    
    # Setting the number of columns for the histogram layout.
    n_cols = 3
    
    # Calculating the number of rows needed for the subplots based on the number of pollutants.
    n_rows = (len(color_map) + n_cols - 1) // n_cols
    
    # Creating a figure and a grid of subplots with specified size.
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 15))
    
    # Flattening the axes array for easy iteration.
    axes = axes.flatten()
    
    # Looping through each pollutant to create a histogram.
    for idx, pollutant in enumerate(color_map.keys()):
        # Creating a histogram for the current pollutant with a kernel density estimate (kde).
        sns.histplot(df[pollutant], color=color_map[pollutant], ax=axes[idx], kde=True)
        
        # Setting the x-label and y-label for each subplot.
        axes[idx].set_xlabel(pollutant)
        axes[idx].set_ylabel('Frequency')

    # Deleting any unused axes if the number of pollutants is less than the number of subplots.
    for j in range(idx + 1, len(axes)):
        fig.delaxes(axes[j])
        
    # Adjusting the layout to make sure the subplots fit well in the figure.
    plt.tight_layout()
    
    # Displaying the created histograms.
    plt.show()

print("City Per Day")
histograms(df_cityday)

print("City Per Hour")
histograms(df_cityhour)

print("Station Per Day")
histograms(df_stationday)

print("Station Per Hour")
histograms(df_stationhour)
```

***Plotting boxplots to decide whether to keep the outliers or not***

```python
# Defining a function 'create_boxplot' that takes in data, a column name, an axes object, and a color.
def create_boxplot(data, column, ax, color):
    # Creating a boxplot for the specified column of the DataFrame on the provided axes.
    # boxprops is used to set the color of the box in the boxplot.
    ax.boxplot(data[column], boxprops=dict(color=color))  
    
    # Setting the title of the boxplot to indicate which column it represents.
    ax.set_title('Boxplot of {}'.format(column))
    
    # Setting the x-label of the boxplot to the column name.
    ax.set_xlabel('{}'.format(column))
    
    # Setting the y-label of the boxplot to indicate that it shows values.
    ax.set_ylabel('Values')

# Creating a list of DataFrames to visualize, which includes city and station data for both daily and hourly measurements.
dataframes = [df_cityday, df_cityhour, df_stationday, df_stationhour]

# Defining a color map to associate specific colors with different pollutants.
color_map = {
    'PM2.5': 'blue', 'PM10': 'green', 'NO': 'red', 'NO2': 'purple',
    'NOx': 'orange', 'NH3': 'brown', 'CO': 'black', 'SO2': 'gray',
    'O3': 'teal', 'Benzene': 'pink', 'Toluene': 'gold', 'Xylene': 'cyan',
    'AQI': 'violet'
}

# Determining the number of plots to create based on the minimum of the number of columns 
# in the first DataFrame and the number of pollutants in the color map.
num_plots = min(len(dataframes[0].columns), len(color_map))

# Creating a figure and a grid of subplots based on the number of DataFrames and the number of plots.
fig, axes = plt.subplots(len(dataframes), num_plots, figsize=(20, 4 * len(dataframes)))

for i, data in enumerate(dataframes):
    for j, (column, color) in enumerate(color_map.items()):
        # Checking if the current index is within the number of plots to be created.
        if j < num_plots: 
            # Calling the create_boxplot function to create a boxplot for the current DataFrame and pollutant.
            create_boxplot(data, column, axes[i, j], color)

plt.tight_layout()

plt.show()
```

***Dropping the columns with less correlation with the target and forming the final features and target dataframe***

```python
# Calculating the correlation of all columns in 'df_corr' with the 'AQI' column.
# This returns a Series where the index is the column names and the values are the correlation coefficients.
corr = df_corr.corr().AQI

# Identifying columns to drop based on their absolute correlation with 'AQI'.
# If the absolute value of the correlation is less than 0.40, the column is considered weakly correlated.
col_to_drop = corr[abs(corr) < 0.40].index

# Creating a new DataFrame 'df_features_reg' by dropping the weakly correlated columns from 'df_final'.
# This helps to retain only those features that have a stronger relationship with 'AQI'.
df_temp = df_final.drop(col_to_drop, axis=1)

# Dropping the 'AQI' and 'AQI_Bucket' columns from 'df_features_reg' - a dataframe for features
df_features_reg = df_temp.drop(['AQI', 'AQI_Bucket'], axis=1)

# Defining the target variable 'df_target_reg' as the 'AQI' column from 'df_final'.
df_target_reg = df_final['AQI']
```

***Since the relation is clearly non-linear and piecewise, let's look for the best model among 'Decision Tree', 'Random Forest' and 'XGB' by performing randomized search and hyperparameter tuning!***

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor as xgb
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler

# Normalize features using MinMaxScaler
scaler = MinMaxScaler()
df_features_reg = scaler.fit_transform(df_features_reg)
df_target_reg = scaler.fit_transform(df_target_reg.values.reshape(-1, 1)).ravel()

# Split data
x_train, x_test, y_train, y_test = train_test_split(df_features_reg, df_target_reg, test_size=0.2, random_state=42)

# Define models and parameter grids
models_params = {
    "DecisionTree": {
        "model": DecisionTreeRegressor(),
        "params": {"max_depth": [3, 5, 7, 10], "min_samples_split": [2, 5, 10]}
    },
    "RandomForest": {
        "model": RandomForestRegressor(random_state=42),
        "params": {"n_estimators": [50, 100, 200], "max_depth": [3, 5, 7], "min_samples_split": [2, 5]}
    },
    "XGBoost": {
        "model": xgb(random_state=42),
        "params": {"n_estimators": [50, 100, 200], "max_depth": [3, 5, 7], "learning_rate": [0.01, 0.1, 0.2]}
    }
}

# Perform RandomizedSearchCV for each model
best_models = {}
for name, model_info in models_params.items():
    grid_search = RandomizedSearchCV(
        model_info["model"], model_info["params"], cv=5, scoring="neg_mean_squared_error", n_jobs=-1, random_state=42
    )
    grid_search.fit(x_train, y_train)

    best_models[name] = {
        "best_estimator": grid_search.best_estimator_,  # Remove the incorrect negative sign
        "best_score": -grid_search.best_score_  # Keep the negative sign only for the score
    }

# Evaluate on test data
test_scores = {}
for name, model_info in best_models.items():
    y_pred = model_info["best_estimator"].predict(x_test)
    test_mse = mean_squared_error(y_test, y_pred)
    test_scores[name] = test_mse
    print(f"{name} - Best MSE on test set: {test_mse:.4f}")

# Find best model
best_model_name = min(test_scores, key=test_scores.get)
print(f"\nBest Model: {best_model_name} with MSE: {test_scores[best_model_name]:.4f}")

# Plot results (Only for first feature)
plt.figure(figsize=(12, 5))
plt.scatter(x_test[:, 0], y_test, color="black", label="Actual Data")

for name, model_info in best_models.items():
    y_pred = model_info["best_estimator"].predict(x_test)
    plt.scatter(x_test[:, 0], y_pred, alpha=0.6, label=f"{name} Predictions")

plt.legend()
plt.title("Comparison of Tree-Based Models with Hyperparameter Tuning")
plt.xlabel("Feature 1 (First Column of X)")
plt.ylabel("Target")
plt.show()
```

```python
import joblib

best_model = best_models[best_model_name]["best_estimator"]  
joblib.dump(best_model, f"aqi_model.joblib")  
print(f"Model saved")
```
